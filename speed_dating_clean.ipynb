```r
# load library
library(dplyr)
```

```r
# import file
speed_dating <- read.csv("Speed Dating Data.csv", stringsAsFactors=FALSE)
```

## Normalize Attributes Code

```r
#to normalize attributes
speed_dating_adj <- speed_dating
df1_1 <- speed_dating[, c("attr1_1", "sinc1_1", "fun1_1", "intel1_1", "amb1_1")]
df1_1
df3_1 <- speed_dating[, c("attr3_1", "sinc3_1", "fun3_1", "intel3_1", "amb3_1")]
df3_1
df3_s <- speed_dating[, c("attr3_s", "sinc3_s", "intel3_s", "fun3_s", "amb3_s")]
df3_s
```

```r
df1_1 <- cbind(df1_1, "sums" = apply(df1_1, 1, sum))
df1_1 <- df1_1[,1:5]/df1_1[,6]*100
#rowSums(df1_1)
```

```r
speed_dating_adj[, c("attr1_1", "sinc1_1", "fun1_1", "intel1_1", "amb1_1")] <- df1_1
#speed_dating_adj[speed_dating$wave >= 6 & speed_dating$wave <= 9, c(70:74)]
```

```r
df3_1 <- cbind(df3_1, "sums" = apply(df3_1, 1, sum))
df3_1 <- df3_1[,1:5]/df3_1[,6]*100
#rowSums(df3_1)
```

```r
speed_dating_adj[, c("attr3_1", "sinc3_1", "fun3_1", "intel3_1", "amb3_1")] <- df3_1
#speed_dating_adj[speed_dating$wave >= 6 & speed_dating$wave <= 9, c(88:92)]
```

```r
df3_s <- cbind(df3_s, "sums" = apply(df3_s, 1, sum))
df3_s <- df3_s[,1:5]/df3_s[,6]*100
#rowSums(df3_s)
```

```r
speed_dating_adj[, c("attr3_s", "sinc3_s", "intel3_s", "fun3_s", "amb3_s")] <- df3_s
#speed_dating_adj[speed_dating$wave >= 6 & speed_dating$wave <= 9, c(115:119)]
```

```r
#adjusted df
speed_dating_adj[, c("attr1_1", "sinc1_1", "fun1_1", "intel1_1", "amb1_1",
                     "attr3_1", "sinc3_1", "fun3_1", "intel3_1", "amb3_1", 
                     "attr3_s", "sinc3_s", "intel3_s", "fun3_s", "amb3_s")]
```

```r
apply(speed_dating_adj[,c("attr1_1", "sinc1_1", "fun1_1", "intel1_1", "amb1_1",
                          "attr3_1", "sinc3_1", "fun3_1", "intel3_1", "amb3_1", 
                          "attr3_s", "sinc3_s", "intel3_s", "fun3_s", "amb3_s")], 2, mean, na.rm = TRUE)
```

```r
summary(speed_dating_adj[,c("attr1_1", "sinc1_1", "fun1_1", "intel1_1", "amb1_1",
                            "attr3_1", "sinc3_1", "fun3_1", "intel3_1", "amb3_1", 
                            "attr3_s", "sinc3_s", "intel3_s", "fun3_s", "amb3_s")])
```

```r
# omitting columns and adding in normalized attributes
speed_dating_clean <- select(speed_dating,-c(field,
                                from,
                                income,
                                go_out,
                                career,
                                exphappy,
                                expnum, 
                                match_es,
                                attr1_s:shar1_s,
                                attr1_s:shar1_s,
                                attr3_s:amb3_s,
                                satis_2,
                                length,
                                numdat_2,
                                attr7_2:shar7_2,
                                attr1_2:shar1_2,
                                attr4_2:shar4_2,
                                attr2_2:shar2_2,
                                attr3_2:amb3_2,
                                attr5_2:amb5_2,
                                attr1_3:shar1_3,
                                attr7_3:shar7_3,
                                attr4_3:shar4_3,
                                attr2_3:shar2_3,
                                attr3_3:amb3_3,
                                attr5_3:amb5_3)) %>%
                     mutate(attr1_1 = speed_dating_adj$attr1_1,
                                sinc1_1 = speed_dating_adj$sinc1_1,
                                fun1_1 = speed_dating_adj$fun1_1,
                                intel1_1 = speed_dating_adj$intel1_1,
                                amb1_1 = speed_dating_adj$amb1_1,
                                attr3_1 = speed_dating_adj$attr3_1,
                                sinc3_1 = speed_dating_adj$sinc3_1,
                                fun3_1 = speed_dating_adj$fun3_1,
                                intel3_1 = speed_dating_adj$intel3_1,
                                amb3_1 = speed_dating_adj$amb3_1,
                                attr3_s = speed_dating_adj$attr3_s,
                                sinc3_s = speed_dating_adj$sinc3_s,
                                fun3_s = speed_dating_adj$fun3_s,
                                intel3_s = speed_dating_adj$intel3_s,
                                amb3_s = speed_dating_adj$amb3_s)
```

```r
write.csv(speed_dating_clean,'speed_dating_clean.csv')
```
